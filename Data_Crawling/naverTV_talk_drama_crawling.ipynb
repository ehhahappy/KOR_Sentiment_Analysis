## **네이버 TV 톡 크롤링(드라마)**

### **사이코지만 괜찮아**

```python
import re
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup 

startpage = int(input())
enddpage = int(input())

List = []
page = startpage
url = "https://tv.naver.com/cjenm.tvNpsycho/talks"
headers = { "referer": url, "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36" } 

while True: 
    c_url = "https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?ticket=ent01&templateId=tvbrandtalk&pool=cbox6&_callback=jQuery111308574130195430252_1598937833743&lang=ko&country=KR&objectId=10393732&categoryId=&pageSize=20&indexSize=5&groupId=&pageType=default&page="+str(page)+"&refresh=false"
    res = requests.get(c_url,headers=headers) 
    cont = BeautifulSoup(res.content, 'html.parser') 
    total_comm = str(cont).split('comment":')[1].split(",")[0] # 실제 댓글 split 
    
    match = re.findall('"contents":([^\*]*),"userIdNo"', str(cont)) # "content": 뒤에 나오는 전체 문장 끝은 ,"userIdNo"로 끝나는...
    List.append(match) 
    
    # 한 번에 댓글이 20개씩 보이기 때문에 한 페이지씩 몽땅 댓글을 긁어오기 
    if int(total_comm) <= ((page*20)): 
        break 
    else: page += 1
        
    if page > enddpage:
        break
        
def flatten(I): 
    flatList = [] 
    for elem in I: 
        if type(elem) == list: #타입이 list이면 그 안의 원소를 추가 
            for e in elem: 
                flatList.append(e)
        else: #타입이 list가 아닌 원소라면 바로 추가 
            flatList.append(elem)
    return flatList 

df_naverTV_talk = pd.DataFrame(np.array(flatten(List)).T)
df_naverTV_talk.to_csv('naverTV_talk_tvNpsycho_{}_{}.csv'.format(startpage, enddpage))
```
